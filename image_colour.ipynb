```python
img = cv2.cvtColor(cv2.imread('Rose.png'), cv2.COLOR_BGR2RGB) 
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
blue, green, red = cv2.split(img)

f1 = plt.figure(figsize = (15, 7.2))
f1.add_subplot(131)
plt.imshow(blue)

f2 = plt.figure(figsize = (15, 7.2))
f2.add_subplot(132)
plt.imshow(green)

f3 = plt.figure(figsize = (15, 7.2))
f3.add_subplot(133)
plt.imshow(red)
```
